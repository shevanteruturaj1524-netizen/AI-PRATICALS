```python
import pandas as pd
import numpy as np
data = {
'Age': [22, 25, 30, 35, 40],
'Salary': [30000, 50000, 60000, 80000, 100000],
'City': ['Delhi', 'Mumbai', 'Delhi', 'Pune', 'Mumbai'],
'Purchased': [0, 1, 0, 1, 1]
}
df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['City_Label'] = le.fit_transform(df['City'])
print(df)
```

```python
df_encoded = pd.get_dummies(df, columns=['City'])
print(df_encoded)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_encoded[['Age', 'Salary']] = scaler.fit_transform(
df_encoded[['Age', 'Salary']]
)
print(df_encoded)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
df_encoded [['Age', 'Salary']] = minmax.fit_transform(
df_encoded [['Age', 'Salary']]
)
print(df_encoded)
```

```python
correlation = df_encoded.corr()
print(correlation['Purchased'])
```

```python
from sklearn.feature_selection import SelectKBest, chi2
X = df_encoded.drop('Purchased', axis=1)
y = df_encoded['Purchased']
selector = SelectKBest(score_func=chi2, k = 2)
X_new = selector.fit_transform(abs(X), y)
print("Selected Features Shape:", X_new.shape)
print(X_new)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, y)
importances = model.feature_importances_
for feature, importance in zip(X.columns, importances):
    print(feature, ":", importance)
```
