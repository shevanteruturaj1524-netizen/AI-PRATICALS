```python
import pandas as pd
import numpy as np
data={
    'Name':['Rahul','Vinit','Himanshu','Ruturaj'],
    'Age':[18,10,np.nan,19],
    'Marks':[100,np.nan,70,50],
    'Branch':['CSE','IT','AIML',None]
     }
```

```python
df=pd.DataFrame(data)
```

```python
print("original Data:\n",df)
```

```python
print("\nMissing Values:\n",df.isnull().sum())
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Marks']=df['Marks'].fillna(df['Marks'].mode()[0])
```

```python
df['Branch']=df['Branch'].fillna('Unknown')
```

```python
df
```

```python
df['Result']=np.where(df['Marks']>=40,'Pass','Fail')
print(df)
```

```python
df.rename(columns={'Marks':'Score'},inplace=True)
print("\nStatistical Summary:\n",df.describe())
```

```python
avg_score=df.groupby('Branch')['Score'].mean()
print("\nAverage Score by Branch:\n",avg_score)
```

```python
high_score=df[df['Score']>80]
print("\nStudents Scoring Above 80:\n",high_score)
```

```python
print("\nMax Score:",np.max(df['Score']))
```

```python
print("Min Score:",np.min(df['Score']))
```

```python
print("Mean Score:",np.mean(df['Score']))
```

```python
print("Standard Deviation:",np.std(df['Score']))
```

```python
print ("\nCleaned and Transformed Data:\n  ",df)
```
