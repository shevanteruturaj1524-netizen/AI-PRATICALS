```python
class Student:
  def __init__(self,name,marks):
    self.name=name
    self.marks=marks

  def display(self):
      print(f"Name:{self.name}")
      print(f"marks:{self.marks}")>

  def result(self):
        return "Pass" if self.marks>=40 else "Fail"

s1=Student("Rasna",85)
s2=Student("Amit",35)
s1.display()
print("Result:",s1.result())
s2.display()
print("Result:",s2.result())
```
